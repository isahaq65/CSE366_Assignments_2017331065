```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Simulatation the analog-analog conversion

**Amplitude Modulation(AM)**<br>
  1.Generate Carrier signal

```python
A=2
f=1
t=np.arange(0, 50, 0.1)

carrier=A*np.sin(2*np.pi*f*t)

plt.figure(figsize=(15, 5))
plt.plot(t, carrier)
plt.show()


```

2.Generate orignal Signal




```python
m = 2.0
A1 = A * m
fm=0.03
phase=np.pi/2

orsig=A1*np.sin((2*np.pi*fm*t)+phase)
plt.figure(figsize=(15, 5))
plt.plot(t, orsig)
plt.show()
```

3.Modulation

```python
y=(1+m*np.sin((2*np.pi*fm*t) + phase))*A*np.sin(2*np.pi*f*t)
plt.figure(figsize=(15, 5))
plt.title('AM Signal')
plt.plot(t, y)
plt.show()
```

<br>**Frequency Modulation(FM)**<br>
1.Generate Carrier signal

```python
A=2
f=.5
t=np.arange(0, 50, 0.1)

carrier=A*np.sin(2*np.pi*f*t)

plt.figure(figsize=(18, 5))

plt.plot(t, carrier)
plt.show()
```

2.Generate original Signal

```python
m = 3.0
A1 = A * m
fm=0.04
phase=np.pi/2

orsig=A1*np.sin((2*np.pi*fm*t)+phase)
plt.figure(figsize=(18, 5))
plt.plot(t, orsig)
plt.show()
```

3.Modulation

```python
y= A*np.sin(2*np.pi * (f * t +A*np.sin(2*np.pi*f*t)))
plt.figure(figsize=(20, 5))
plt.title('FM Signal')
plt.plot(t, y)
plt.show()
```

<br>**Phase Modulation(PM)**<br>
1.Generate original Signal

```python
A=2
f=1
t=np.arange(0, 50, 0.1)

carrier=A*np.sin(2*np.pi*f*t)

plt.figure(figsize=(18, 5))

plt.plot(t, carrier)
plt.show()
```

2.Generate original Signal

```python
m = 4.0
A1 = A * m
fm=0.06
phase=np.pi/2
ampSen=2;

orsig=A1*np.sin((2*np.pi*fm*t)+phase)
plt.figure(figsize=(18, 5))
plt.plot(t, orsig)
plt.show()
```

3.Modulation

```python
y=A*np.sin(2*np.pi*f*t) + (ampSen * orsig + phase)

plt.figure(figsize=(20, 5))
plt.title('PM Signal')
plt.plot(t, y)
plt.show()

```
