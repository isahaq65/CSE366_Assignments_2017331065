```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 300)
```

$$x(t)= A cos(2\pi ft+ @)$$

```python
#fuction for generate signal

def generate_sinusoidal_signal(t, A, f, initial_phase):
  x = A * np.cos(2.0 * np.pi * f * t + initial_phase)
  return x
```

```python
#3 different signal 

y1= generate_sinusoidal_signal(t, 2, 1, 10)
y2= generate_sinusoidal_signal(t, 2, 4, 6)
y3= generate_sinusoidal_signal(t, 2, 8, 12)
```

```python
fig, position = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
fig.suptitle('Three signal with different properties and  one Composite Signal ', fontsize=15)

position[0, 0].plot(t,y1)
position[0, 0].set_title('signal (1)', fontsize = 15)

position[0, 1].plot(t,y2, 'r')
position[0, 1].set_title('signal (2)', fontsize = 15)

position[1,0].plot(t,y3,'g')
position[1,0].set_title('signal (3)', fontsize = 15)

position[-1,-1].plot(t, y1 + y2 + y3)
position[-1,-1].set_title('Composite Signal', fontsize = 15)



plt.show()
```

**Sampling**

```python
#fs - sampling frequency that must be greater than two times of original signal frequency 
#N- number of step

fs = 100 
N = 1.0/fs
n = np.arange(0, 1.0, N)
x1 =  generate_sinusoidal_signal(n, 1, 5, 0)
x2 =generate_sinusoidal_signal(n, 1, 10,0) 
x3= generate_sinusoidal_signal(n, 1, 15,0)
plt.title('Sampling frequency= 10', fontsize = 20)
plt.stem(n,x1+x2+x3)

plt.show()



fs = 80 
N = 1.0/fs
n = np.arange(0, 1.0, N)
x1 =  generate_sinusoidal_signal(n, 1, 5, 0)
x2 =generate_sinusoidal_signal(n, 1, 10,0) 
x3= generate_sinusoidal_signal(n, 1, 15,0)
plt.title('Sampling frequency= 80', fontsize = 20)
plt.stem(n,x1+x2+x3)

plt.show()

fs = 20 
N = 1.0/fs
n = np.arange(0, 1.0, N)
x1 =  generate_sinusoidal_signal(n, 1, 5, 0)
x2 =generate_sinusoidal_signal(n, 1, 10,0) 
x3= generate_sinusoidal_signal(n, 1, 15,0)
plt.title('Sampling frequency= 10', fontsize = 20)
plt.stem(n,x1+x2+x3)

plt.show()


```
