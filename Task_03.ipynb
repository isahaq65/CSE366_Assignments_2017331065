```python
import matplotlib.pyplot as plt
import numpy as np
```

##simulation the digital-analog conversion 

**unipolar nrz encoding**

```python
def unipolar_nrz_encoding(bits,N,bitrate,T,t):
    y= np.array([])
    for i in range(N):
        tmp =bits[i]* np.ones((T))
        y= np.append(y,tmp)
    ts = np.linspace(0,N*t,len(y))
    return [ts ,y]

def plot(ts,y):
    plt.figure(figsize=(10, 10))
    plt.plot(ts,y)
    plt.show()
```

**ASK Conversion**

```python
bits = np.array([1, 0, 1, 1, 0, 1, 0, 0, 1, 0, 1])
t=10e-6
bitrate=1/t
N=len(bits)
T=100
tx,y=unipolar_nrz_encoding(bits,N,bitrate,T,t)
plot(tx,y)

ac = 7                        
fc =round(15*bitrate)
tc = np.arange(0,N*t,t/T)
xc = ac * np.sin(2.0 * np.pi * fc * tc)
plot(tc,xc)

print(len(y))
print(len(xc))
ms= y * xc
print(ms)
plot(tc,ms)
```

**FSK Conversion**

```python
A = 16
f1 =round(10*bitrate)
f2 =round(20*bitrate)
tc = np.arange(0, t, t/T)
fsk = np.array([])

for i in range(N):
  if bits[i] == 1:
    sig = A * np.sin(2.0 * np.pi * f1 * tc)
    fsk = np.append(fsk, sig)
  else:
    sig = A * np.sin(2.0 * np.pi * f2 * tc)
    fsk = np.append(fsk, sig)
plot(tx,fsk)
```

**PSK Conversion**

```python
A = 11                      
fc =round(7*bitrate)
phase1 = 0
phase2 = np.pi
tc = np.arange(0, t, t/T)
psk_ms = np.array([])
for i in range(N):
  if bits[i] == 1:
    sig = A * np.sin((2.0 * np.pi * fc * tc) + phase1)
    psk_ms = np.append(psk_ms, sig)
  else:
    sig = A * np.sin((2.0 * np.pi * fc * tc) + phase2)
    psk_ms = np.append(psk_ms, sig)
plot(tx, psk_ms)

```
